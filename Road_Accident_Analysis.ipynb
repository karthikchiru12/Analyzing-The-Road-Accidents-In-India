## 1. Importing the required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from shapely.geometry import Point, Polygon
import gmplot
```

## 2. About the dataset

<blockquote>This dataset contains information of road accidents from 2000-2012.</blockquote>

### 2.1 Loading the dataset

```python
data = pd.read_excel('DATA/datafile.xls')
data.head()
```

### 2.2 Objective

<blockquote>Analyzing the Road accidents in our country.</blockquote>

## 3. Understanding Various High Level Information About The Dataset

```python
print("Shape of the dataset   :",data.shape)
print("\nColumns in the dataset :",list(data.columns))
print("\n\nMissing Values         : No missing values\n")
print(data.isnull().sum())
print("\nOther information      :\n")
_info=pd.DataFrame(data.describe().T)
_info
```

## 4. Segregating the data by states

### 4.1 Getting list of states

```python
states_list=list(data['States'].unique())
print(states_list)
```

### 4.2 Making new dataframes for each state

```python
Andhra_pradesh=data[data.States=='ANDHRA PRADESH']
Arunachal_pradesh=data[data.States=='ARUNACHAL PRADESH']
Assam=data[data.States=='ASSAM']
Bihar=data[data.States=='BIHAR']
Chhattisgarh=data[data.States=='CHHATTISGARH']
Goa=data[data.States=='GOA']
Gujarat=data[data.States=='GUJARAT']
haryana=data[data.States=='HARYANA']
Himachal_pradesh=data[data.States=='HIMACHAL PRADESH']
Jammu_kashmir=data[data.States=='JAMMU & KASHMIR']
Jharkhand=data[data.States=='JHARKHAND']
Karnataka=data[data.States=='KARNATAKA']
Kerala=data[data.States=='KERALA']
Madhya_pradesh=data[data.States=='MADHYA PRADESH']
Maharasthra=data[data.States=='MAHARASHTRA']
Manipur=data[data.States=='MANIPUR']
Meghalaya=data[data.States=='MEGHALAYA']
Mizoram=data[data.States=='MIZORAM']
Nagaland=data[data.States=='NAGALAND']
Odisha=data[data.States=='ODISHA']
Punjab=data[data.States=='PUNJAB']
Rajasthan=data[data.States=='Rajasthan']
Sikkim=data[data.States=='SIKKIM']
Tamil_nadu=data[data.States=='TAMIL NADU']
Tripura=data[data.States=='TRIPURA']
Uttar_pradesh=data[data.States=='UTTAR PRADESH']
Uttarakhand=data[data.States=='UTTARAKHAND']
West_bengal=data[data.States=='WEST BENGAL']
Andaman_and_nicobar=data[data.States=='Andaman And Nicobar']
Chandigarh=data[data.States=='CHANDIGARH']
Dadra_and_nagar_haveli=data[data.States=='Dadra And Nagar Haveli']
Daman_and_diu=data[data.States=='DAMAN & DIU']
Delhi=data[data.States=='Delhi']
Lakshadweep=data[data.States=='LAKSHADWEEP']
Pudducherry=data[data.States=='PUDDUCHERRY']
```

## 5. Analysis on the data

### 5.1 Number of males who died in the road accidents with their causes

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Male", y="CAUSE", data=data)
plt.title("Causes v/s Number of males who died in Road Accidents")
```

<blockquote>
    <ul>
        <li>Most of the people died by Truck/Lorry and Two wheeler.</li>
        <li>Least number of people died by Tempo/Vans.</li>
        <li>Other than the Bus, No other government vehicles caused a lot of death.</li>
    </ul>
</blockquote>

### 5.2 Number of females who died in the road accidents with their causes

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Female", y="CAUSE", data=data)
plt.title("Causes v/s Number of females who died in Road Accidents")
```

<blockquote>
    <ul>
        <li>Relativelty there were more female deaths by private bus.</li>
        <li>Observations made on male deaths satisfy in here too.</li>
    </ul>
</blockquote>

### 5.3 Both male and female who died in the road accidents with their causes

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Total", y="CAUSE", data=data)
plt.title("Causes v/s Number of people who died in Road Accidents")
```

### 5.4 Number of people died in road accidents in each year

```python
sns.barplot(x="Year", y="Total", data=data)
plt.title("Year v/s Number of people who died in Road Accidents")
```

<blockquote>
    <ul>
        <li>Most of the deaths happened in the year 2012.</li>
        <li>Least number of people died in the year 2001.</li>
    </ul>
</blockquote>

### 5.5 States v/s Number of people died in road accidents

```python
sns.set(rc={'figure.figsize':(20,18)})
sns.barplot(x="Total", y="States", data=data)
plt.title("Year v/s Number of people died in road accidents")
```

<blockquote>
    <ul>
        <li>There are huge number of people from Andhra Pradesh who died in road accidents.</li>  
        <li>And followed by Maharasthra, Tamilnadu and Uttar pradesh.</li>
        <li>Less number of people died in Nagaland,Mizoram,Lakshwadeep,Daman and diu, Andaman and nicobar,Sikkim,Meghalaya,Manipur,Goa and Arunachal Pradesh</li>
   </ul>
</blockquote>

### 5.6 Heatmap for various features

### 5.7 Creating a map of indian roads and plotting the data

```python
# Building an indian roads map from shape file
PATH='/home/chiranjeevi_karthik/Desktop/Accident analysis/Indian Roads Shape FIles/IND_rds/IND_roads.shp'
indian_roads = gpd.read_file(PATH)
crs={'init':'espg:7755'}
cords= [Point(xy) for xy in zip(data['Lat'],data['Long'])]
gdf=gpd.GeoDataFrame(data,crs=crs,geometry=cords)
fig,ax=plt.subplots(figsize=(15,15))
indian_roads.plot(ax=ax,alpha=0.4,color='green')
#gdf[gdf['States']=='ANDHRA PRADESH'].plot(ax=ax,markersize=1,color="red",label="AP")
```

## 6. Statewise Analysis of Road Accidents

### 6.1 Andhra pradesh

#### 6.1.1 Number of Males who died every year in Andhra pradesh Road Accidents

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x="Year", y="Male",data=Andhra_pradesh)
plt.title("Number of Males who died every year in Andhra pradesh")
```

<blockquote>
    <ul>
        <li>The Number of Males who died in Andhra pradesh linealy increased from 2002 to 2012.</li>

#### 6.1.2 Distribution of males who died in Andhra Pradesh Road Accidents

```python
sns.FacetGrid(Andhra_pradesh, height=6).map(sns.distplot, "Male").add_legend()
plt.title("Distribution of males who died in Andhra Pradesh")
```

#### 6.1.3 Number of females who died every year in Andhra Pradesh Road Accidents

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x="Year", y="Female",data=Andhra_pradesh)
plt.title(" Number of females who died every year")
```

<blockquote>
    <ul>
        <li>There are relatively less females died in Andhra pradesh than males.</li>
    </ul>
</blockquote>

#### 6.1.4 Distribution of females who died in Andhra Pradesh Road Accidents

```python
sns.FacetGrid(Andhra_pradesh, height=8).map(sns.distplot, "Female").add_legend()
plt.title("Distribution of females who died in Andhra Pradesh")
```

#### 6.1.5 Causes v/s Number of males who died in Andhra Pradesh Road Accidents

### 6.1.7 Map of Accidents For Andhra Pradesh
